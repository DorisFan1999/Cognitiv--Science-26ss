### Exercise 5 (Strong Law of Large Numbers)

#### Task 1

Simulate random samples ${x_i}^n_i=1$ of different sizes $n \in {1,2,...,1000}$ from any probability distribution of your choice. Then compute the sample mean given by:

$$ \bar{x}_n = \frac{1}{n} \sum_{i=1}^n x_i$$

for all samples of different sizes. Plot the value of the mean (on the $y$-axis) w.r.t. the number of samples (on the $x$-axis). Do the means converge to a given value with increasing value of $n$? Explain your results in brief.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Choose normal distribution where X ~ N(3, 1)
# the theoretical mean is 3

seed = np.random.default_rng(seed=42)
n = 1000
mu = 3
sigma = 1

samples = seed.normal(mu, sigma, size=n)
sample_sizes = np.arange(1, n + 1)
sample_means = np.cumsum(samples) / sample_sizes

# plot
plt.figure(figsize=(6, 3.5))
plt.plot(sample_sizes, sample_means, label="sample mean", alpha=0.8, linewidth=1)
plt.axhline(mu, color="red", linestyle="--", label="theoretical mean = 3")
plt.xlabel("Number of samples")
plt.ylabel("Sample mean")
plt.title("Convergence of Sample Mean")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Sample mean at n={n}: {sample_means[-1]:.4f}")
```

From the result of shown, we can know that the law of large numbers guarantees the convergence of the sample. When $n$ is sufficiently large, the sample mean converges to theoretical mean with increasing value of $n$. In this example, we choose Normal distribution where $X$ ~ $N(3, 1)$, and as we can see, after 1000 samples, the sample mean is approximately 2.9711, which is very close to the theoretical value.
